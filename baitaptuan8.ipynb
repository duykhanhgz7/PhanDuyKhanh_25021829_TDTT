```python
# W7A16
arr = list(map(int, input().split()))
ssa = sorted(set(arr))

quan = 0
flag = True
for i in ssa:
    freq = arr.count(i)
    if freq == i:
        pass
    else:
        if freq < i:
            for j in range(freq):
                arr.remove(i)
            quan += freq
        else:
            for j in range(0, freq - i):
                arr.remove(i)
            quan += freq - i
        flag = False
if flag:
    print('YES')
else:
    print('NO', end = ' ')
    print(quan)
```

```python
# 25
numbers = list(input())
a = numbers.copy()
for i in range(len(a)):
    a[i] = int(a[i])
k = int(input())
findnum = []
if len(a) == k:
    print(''.join(numbers))
else:
    start = 0
    for i in range(k-1, -1, -1):
        findnum.append(max(a[start:(len(a)-i)]))
        b = max(a[start:(len(a)-i)])
        for j in range(start, len(a)):
            if a[j] == b:
                start = j + 1
                break
    for i in range(len(findnum)):
        findnum[i] = str(findnum[i])
    print(''.join(findnum))
```

```python
# 28
def prime_number(x):
    for i in range(1, int(x**0.5 + 1)):
        if x % i == 0 and i != 1 :
            return False
    return True
N = int(input())
numbers = []
for i in range(1, N + 1):
    if not prime_number(i):
        numbers.append(i)
if len(numbers) == 1:
    print(numbers[0])
else:
    ca = []
    for i in range(0, len(numbers)):
        sca = [numbers[i]]
        ca.append(sca)
        if i > 0:
            if numbers[i] - numbers[i-1] == 1:
                for j in reversed(ca[i-1]):
                    ca[i].insert(0, j)
    maxsca = 0
    for i in ca:
        if len(i) > maxsca:
            maxsca = len(i)

    for i in ca:
        if len(i) == maxsca:
            for j in range(0, len(i)):
                if i[j] != N:
                    if j + 1 != len(i):
                        print('{},'.format(i[j]), end = ' ')
                    else:
                        print(i[len(i)-1])
            break
```

```python
# 29 
arr = list(map(int, input().split()))

if len(arr) == 1:
    print(arr[0])
else:
    lcca = []
    for i in range(len(arr)):
        cca = [arr[i]]
        lcca.append(cca)
        if i > 0:
            if arr[i] > arr[i - 1]:
                for j in range(len(lcca[i-1])-1, -1, -1):
                    lcca[i].insert(0, lcca[i-1][j])
    lma = 0
    for i in lcca:
        if len(i) > lma:
            lma = len(i)
    for i in lcca:
        if len(i) == lma:
            for j in i:
                print(j, end = ' ')
            break
```

```python
# 30 
arr = list(map(int, input().split()))

if len(arr) == 1:
    print(arr[0])
else:
    lcanc = []
    lcancn = []
    for i in range(len(arr)):
        canc = [arr[i]]
        step = 1
        lcancn.append(step)
        lcanc.append(canc)
        k = 0
        for j in range(0, i):
            if arr[i] > arr[j] and lcancn[j] > k:
                k = lcancn[j]
                lcancn[i] = lcancn[j] + 1
        for j in range(0, i):
            if arr[i] > arr[j] and lcancn[j] == k:
                for t in lcanc[j][::-1]:
                    lcanc[i].insert(0, t)
                break
    lg = 0
    for i in lcanc:
        if len(i) > lg:
            lg = len(i)
    for i in lcanc:
        if len(i) == lg:
            for j in i:
                print(j, end = ' ')
            break
```
