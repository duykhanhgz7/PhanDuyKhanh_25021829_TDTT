```python
def Max_of_Two(a , b):
    if a > b:
        return a
    else:
        return b
Max_of_Two(10, 20)
```

```python
def is_prime_number(x):
    for i in range(1, int(x**0.5) + 1):
        if x % i == 0 and i != 1:
            return print("x khong la so nguyen to")
            break
    else:
        return print("x la so nguyen to")
is_prime_number(10)
```

```python
def is_pfnum(x):
    s = 1
    for i in range(1, int(x**0.5) + 1):
        if x % i == 0 and x / i != x:
            if x % i == i:
                s += i
            else:
                s += i
                s += x/i
    return s == x
is_pfnum(6)
```

```python
def coor(lst, k):
    for i in lst:
        if i == k:
            return lst.index(i)
    return -1
coor([1, 2, 3, 4, 5], 5)
```

```python
def factorial(x):
    s = 1
    for i in range(1, x + 1):
        s *= i
    return s
factorial(5)
```

```python
def cal(n1, n2, op):
    match(op):
        case '+':
            return round(n1 + n2, 2)
        case '-':
            return round(n1 - n2, 2)
        case '*':
            return round(n1 * n2, 2)
        case '/':
            return round(n1 / n2, 2)
cal(1, 3, '/')    
```

```python
def sum_ele(x):
    nums = list(str(x))
    s = 0
    for i in nums:
        s += int(i)
    return s
sum_ele(123)
```

```python
def dangcau(a,b):
    quytac = len(set(zip(a,b)))
    if quytac == len(set(a)) == len(set(b)):
        return True
    return False
dangcau("abbz", "xybz")
```
