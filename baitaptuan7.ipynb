```python
# W7A1
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
arr = list(map(int, input().split()))
target = int(input())
binary_search(arr, target)
```

```python
# W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
arr = list(map(int, input().split()))

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
print(selection_sort(arr))
```

```python
# W7A4
def appear_most(arr):
    dic = {}
    for i in arr:
        if i not in dic.keys():
            dic[i] = arr.count(i)
    a = 0
    x = 0
    for i in dic.keys():
        if dic[i] > a:
            x = i
            a = dic[i]
        if dic[i] == a:
            continue
    return x
arr = list(map(int, input().split()))
appear_most(arr)
```

```python
# W7A6
def flat_list(arr):
    nothing = []
    for i in arr:
        nothing += i
    return nothing
arr = eval(input())
print(flat_list(arr))
```

```python
# W7A7
def child_rise(arr):
    a = []
    for i in range(0, len(arr)):
        step = 1
        a.append(step)
        if i > 0:
            x = 0
            for j in range(0, i):
                if arr[i] > arr[j] and a[j] > x:
                    step = a[j] + 1
                    x = a[j]
            a[i] = step
    return max(a)
arr = list(map(int, input().split()))
child_rise(arr)      
```

```python
# W7A8
interval = eval(input())
queries = list(map(int, input().split()))
dapans = []
for i in range(len(queries)):
    dapans.append(-1)
    length = 10**8
    for j in range(len(interval)):
        if interval[j][0] <= queries[i] <= interval[j][1]:
            if interval[j][1] - interval[j][0] < length:
                length = interval[j][1] - interval[j][0]
                dapans[i] = length
print(dapans)
```

```python
# W7A10
arr = list(map(int, input().split()))
n = int(input())
ml = []
for i in range(len(arr)):
    mod = 0
    for j in range(i + 1, len(arr)):
        a = arr[i:j+1]
        if sum(a) % n > mod:
            mod = sum(a) % n
    ml.append(mod)
print(max(ml))
```
