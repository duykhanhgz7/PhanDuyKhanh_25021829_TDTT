```python
#W6A1
x = list(map(int, input().split()))
y = []
for i in x:
    if i not in y:
        y.append(i)
print(y)
```

```python
#W6A2
x = list(map(int, input().split()))
tong = []
tong.append(x[0])
for i in range(1, len(x)):
    congdon = sum(x[:(i+1)])
    tong.append(congdon)
print(tong)
```
